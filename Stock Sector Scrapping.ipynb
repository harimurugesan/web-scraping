```python
import datetime
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
def scrapAtags(link, sector) :
    current_response = requests.get(link)
    current_soup = BeautifulSoup(current_response.text,'html.parser')
    index = 0
    for tables in current_soup.find_all('table', attrs={'class': 'tbldata14 bdrtpg'}) :
        for td in tables.find_all('td', attrs={'class': 'brdrgtgry'}) :
             for a_tag in td.find_all('a', href=True):
                    if a_tag['href'].startswith('/india/stockpricequote') :
                        final_response = requests.get("http://www.moneycontrol.com"+a_tag['href'])
                        final_soup = BeautifulSoup(final_response.text,'html.parser')
                        for divs in final_soup.find_all('div', attrs={'class': 'FL gry10'}) :
                            final_res = {}
                            final_text = (divs.text).split('|')
                            for ele in final_text :
                               ele = re.sub('[\n\r\t]', '', ele)
                               ele = ele.split(':')
                               if len(ele) == 2 :
                                   final_res[ele[0].strip()] = ele[1].strip()
                            if final_res :
                                final_res['SUB SECTOR'] = final_res['SECTOR']
                                final_res['SECTOR'] = sector
                                finalArray.append(final_res)
    print(finalArray)
```

```python
path = "http://www.moneycontrol.com/india/stockmarket/sector-classification/marketstatistics/nse/automotive.html"
res = requests.get(path)
soup = BeautifulSoup(res.text,'html.parser')
sectors = []
sectorLinks = []
sectorStocks = {}
finalArray = []

## Get all the sectors
for div in soup.find_all('div', attrs={'class': 'lftmenu'}) :
    for li in div.find_all('li'):
        for a_tag in li.find_all('a', href=False) :
            if a_tag.text not in sectors:
                print("RUNNING FOR SECTOR: ", a_tag.text)
                scrapAtags(path, a_tag.text)
                sectors.append(a_tag.text)
                sectorLinks.append(path)
        for a_tag in li.find_all('a', href=True) :
            sectors.append(a_tag.text)
            sectorLinks.append("http://www.moneycontrol.com"+a_tag['href'])
            print("RUNNING FOR SECTOR: ", a_tag.text)
            scrapAtags("http://www.moneycontrol.com"+a_tag['href'], a_tag.text)
        
```
