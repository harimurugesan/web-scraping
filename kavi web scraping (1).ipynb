```python
#  edit the range and final path
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
col = ['Name of the Institute', 'Short Name', 'Reference ID', 'Address', 'District', 'City', 'Telephone 1', 'Telephone 2', 'Fax', 'WebSite', 'Email', 'Landmark', 'Year Of Starting', 'Institute Minority', 'Name of the Management/Trust', 'Correspondence/Chairman Name', 'Name of Principal & Qualification', 'Building Space for Instructional Area', 'Address of Administrative Office', 'Head Office Telephone 1', 'Head Office Telephone 2', 'Head Office Fax']
result = []
#200923,201032
for i in range(200923,201032):
    res = requests.get("http://www.ragingbulledu.com/Default_InstProfile.aspx?instid={}".format(i))
    soup = BeautifulSoup(res.text,'html.parser')
    li = []
    for i in soup.find_all('tr'):
        for j in  i.findAll('td'):
            for k in  j.findAll('span',attrs={"class":"BlackFont12"}):
                if (":" in i.text) and (i.text not in li):
                    li.append(i.text)
                #print(li)
    result.append([i.split(':')[1].replace("\n","unknown").replace("unknown","").replace(",,","") for i in li[2:]])
pd.DataFrame(result,columns=col).to_csv(r"/home/clustrex-22/college_details.csv",index= False)
```

```python
import requests
import pandas as pd
limit = list(range(0,2000,100))
for year in range(2011,2020):
    result_df = pd.DataFrame()
    for li in limit:
        #print(year,li)
        r = requests.get("https://www.forbes.com/forbesapi/source/more.json?filter=&limit=2000&listYear={}&retrievedFields\
=&sort=&source=list&sourceType=organization&sourceValue=global2000&start={}".format(year,li))
        df = pd.DataFrame(r.json()['promotedContent']['contentPositions'])
        result_df = result_df.append(df)
        #print(result_df.shape)
    result_df.to_csv("/home/clustrex-22/Downloads/forbes_2000/forbes_{}.csv".format(year),index=False)
```
